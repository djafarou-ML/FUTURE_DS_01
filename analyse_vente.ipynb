```python
#importation des librairies
import numpy as np
import pandas as pd
import missingno as mson
import matplotlib.pyplot as plt
import seaborn as sn 
```

```python
# Chargement du dataset Online Retail et affichage des 10 premières lignes
df = pd.read_csv('online_retail.csv')
df.head(10)
```

```python
#Affichage des 10 denières lignes
df.tail(10)
```

```python
# Informations générales sur le DataFrame
df.info()
```

```python
#visualisation des valeurs manquantes
mson.matrix(df)
plt.show()
```

```python
# correlation entre lles valeures manquantes
mson.dendrogram(df)
plt.show()
```

```python
# Statistiques des variables numériques
df.describe()
```

```python
# Comptage du nombre total de valeurs manquantes par colonne
data_absent = df.isnull().sum()
data_absent
```

```python
#Identification des CustomerID manquantes
customer_absent = df[df['CustomerID'].isnull()]
customer_absent
```

```python
# Identification des doublons
duplicate = df[df.duplicated()]
duplicate.head(10)
```

```python
# Identification des lignes où le prix unitaire est négatif
negative_unit_price = df.loc[df['UnitPrice']  < 0]
negative_unit_price
```

```python
# Identification des prix efale à zéro
price_zero = df[df['UnitPrice'] ==0]
price_zero
```

```python
#Identification des quantités négatives
negative_quantity = df.loc[df['Quantity']  < 0]
negative_quantity
```

```python
'''Creation de DataFrame contenant les mauvaises donnes : commande annuler, quantite negatif et prix et perte
    et regroupement de toutes les données problématiques dans un seul DataFrame
'''
bad_data = pd.concat([customer_absent, negative_unit_price, negative_quantity, duplicate, price_zero])
bad_data
```

```python
# Suppression des lignes identifiées comme problématiques à partir de leurs index
df = df.drop(bad_data.index, errors='ignore')
df
```

```python
# Conversion de 'InvoiceDate' en datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df
```

```python
df_clean = df.copy()
```

```python
# Affichage du DataFrame après nettoyage
df_clean.isnull().sum()
```

```python
# Vérification des types et du nombre de valeurs restantes après suppression
df_clean.info()
```

```python
# Vérification des statistiques après nettoyage
df_clean.describe()
```

```python
# Calcul du revenu pour chaque ligne
df_clean['Revenu'] = df_clean['Quantity'] * df_clean['UnitPrice']
df_clean
```

```python
# Calcul des 20 produits les plus rentables
top_products = df_clean.groupby('Description')['Revenu'].sum().sort_values(ascending=False)[:20]
top_products.head(15)
```

```python
# Création du graphique pour le Top 20 des produits les plus rentables

plt.figure()
plt.subplot(111)

# Récupération des labels et valeurs

labels_X = top_products.index
valeurs_Y = top_products.values

plt.bar(labels_X, valeurs_Y)
plt.xticks(rotation=90)

# Ajout des labels et titre

plt.ylabel("Revenu total")
plt.title("Top 20 des produits générant le plus de revenu")

plt.show()
```

```python
# Extraire le numéro de la semaine à partir de la date de la facture

df_clean['Weeks'] = df_clean['InvoiceDate'].dt.isocalendar().week
df_clean
```

```python
# Calcul du revenu total pour chaque semaine

top_products_week = df_clean.groupby('Weeks')['Revenu'].sum()
top_products_week
```

```python
sn.set_context("notebook", rc={"xtick.labelsize": 15, "ytick.labelsize": 15})
# Trier les semaines par revenu décroissant et garder les 20 meilleures

top_products_week = top_products_week.sort_values(ascending=False)
top_products_week = top_products_week[:20]


labels_X = top_products_week.index
valeurs_Y = top_products_week.values

top_dates_df = pd.DataFrame({'Weeks': labels_X, 'Revenu': valeurs_Y})

plt.figure(figsize=(15, 6))
plt.subplot(111)
sn.pointplot(data=top_dates_df, x='Weeks', y='Revenu')
plt.xticks(rotation=90)

plt.ylabel("Revenu")
plt.title("Top des semaines générant le plus de revenu")

plt.show()
```

```python
# Extraire la période à partir de la date de la facture

df_clean['Mois'] = df_clean['InvoiceDate'].dt.to_period('M')
df_clean
```

```python
# Calcul du revenu total pour chaque mois
top_products_mois = df_clean.groupby('Mois')['Revenu'].sum()
top_products_mois
```

```python
sn.set_context("notebook", rc={"xtick.labelsize": 15, "ytick.labelsize": 15})

top_products_mois = top_products_mois.sort_values(ascending=False)
top_products_mois = top_products_mois[:20]

labels_X = top_products_mois.index
valeurs_Y = top_products_mois.values

top_dates_df = pd.DataFrame({'Mois': labels_X, 'Revenu': valeurs_Y})

plt.figure(figsize=(15, 6))
plt.subplot(111)

sn.pointplot(data=top_dates_df, x='Mois', y='Revenu')

plt.xticks(rotation=90)

plt.ylabel("Revenu")
plt.title("Top des mois générant le plus de revenu")

plt.show()
```

```python
sn.set_context("notebook", rc={"xtick.labelsize": 15, "ytick.labelsize": 15})

top_dates = top_dates.sort_values(ascending=False)
top_dates = top_dates[:20]

labels_X = top_dates.index
valeurs_Y = top_dates.values

top_dates_df = pd.DataFrame({'InvoiceDate': labels_X, 'Revenu': valeurs_Y})

plt.figure(figsize=(15, 6))
plt.subplot(111)

sn.pointplot(data=top_dates_df, x='InvoiceDate', y='Revenu')

plt.xticks(rotation=90)

plt.ylabel("Revenu")
plt.title("Top 20 des dates générant le plus de revenu")

plt.show()
```
