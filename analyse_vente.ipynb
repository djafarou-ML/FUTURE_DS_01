```python
# Import libraries
import numpy as np
import pandas as pd
import missingno as mson
import matplotlib.pyplot as plt
import seaborn as sn 
```

```python
# Load Dataset
df = pd.read_csv('online_retail.csv')
df.head(10)
```

```python
df.tail(10)
```

```python
df.info()
```

```python
# Visualize missing values
mson.matrix(df)
plt.show()
```

```python
# Correlation between missing values
mson.dendrogram(df)
plt.show()
```

```python
# Statistics of numeric variables
df.describe()
```

```python
# Count total missing values per column
missing_data = df.isnull().sum()
missing_data
```

```python
# Identify missing CustomerID
customer_missing = df[df['CustomerID'].isnull()]
customer_missing
```

```python
# Identify duplicate rows
duplicate = df[df.duplicated()]
duplicate.head(10)
```

```python
# Identify rows with negative unit price
negative_unit_price = df.loc[df['UnitPrice'] < 0]
negative_unit_price
```

```python
# Identify rows with zero unit price
price_zero = df[df['UnitPrice'] == 0]
price_zero
```

```python
# Identify rows with negative quantity
negative_quantity = df.loc[df['Quantity'] < 0]
negative_quantity
```

```python
'''
Create a DataFrame containing problematic data:
canceled orders, negative quantities, negative or zero prices, duplicates.
Combine all problematic data into a single DataFrame.
'''
bad_data = pd.concat([customer_missing, negative_unit_price, negative_quantity, duplicate, price_zero])
bad_data
```

```python
# Remove rows identified as problematic by their index
df = df.drop(bad_data.index, errors='ignore')
df
```

```python
# Strip extra spaces in 'Country' column
df['Country'] = df['Country'].str.strip()
df
```

```python
# Convert 'InvoiceDate' to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df
```

```python
# Create a cleaned copy of the DataFrame
df_clean = df.copy()
```

```python
# Check for missing values after cleaning
df_clean.isnull().sum()
```

```python
# Verify types and remaining values after removal
df_clean.info()
```

```python
# Check statistics after cleaning
df_clean.describe()
```

```python
# Calculate revenue per row
df_clean['Revenue'] = df_clean['Quantity'] * df_clean['UnitPrice']
df_clean
```

```python
# Top 20 most profitable products
top_products = df_clean.groupby('Description')['Revenue'].sum().sort_values(ascending=False)[:20]
top_products
```

Data Cleaning

The dataset initially contained missing CustomerIDs, duplicate transactions, negative quantities, and zero or negative prices.
These records likely correspond to canceled orders or data entry errors.
After cleaning, df_clean contains only valid transactions, ensuring more reliable revenue analysis.

```python
# Plot Top 20 most profitable products
plt.figure()
plt.subplot(111)

# Get labels and values
labels_X = top_products.index
values_Y = top_products.values

plt.bar(labels_X, values_Y)
plt.xticks(rotation=90)

# Add labels and title
plt.ylabel("Total Revenue")
plt.title("Top 20 Products Generating the Most Revenue")

plt.show()
```

Top Products

A small number of products generate a significant share of total revenue.
This indicates revenue concentration on best-performing items.
The company should prioritize inventory management and marketing campaigns for these high-performing products to maximize profitability.

```python
# Extract week number from InvoiceDate
df_clean['Weeks'] = df_clean['InvoiceDate'].dt.isocalendar().week
df_clean
```

```python
# Calculate total revenue per week
top_products_week = df_clean.groupby('Weeks')['Revenue'].sum()
top_products_week
```

```python
# Plot revenue per week
sn.set_context("notebook", rc={"xtick.labelsize": 15, "ytick.labelsize": 15})
top_products_week = top_products_week.sort_index()

labels_X = top_products_week.index
values_Y = top_products_week.values
top_dates_df = pd.DataFrame({'Weeks': labels_X, 'Revenue': values_Y})

plt.figure(figsize=(15, 6))
plt.subplot(111)
sn.pointplot(data=top_dates_df, x='Weeks', y='Revenue')
plt.xticks(rotation=90)

plt.ylabel("Revenue")
plt.title("Weeks Generating the Most Revenue")

plt.show()
```

Weekly Trend

Weekly revenue shows fluctuations that reflect short-term variations in sales performance.
Monitoring weekly trends helps detect unusual spikes or declines early, enabling faster operational adjustments.g targeted marketing campaigns in advance.

```python
# Extract month from InvoiceDate
df_clean['Month'] = df_clean['InvoiceDate'].dt.to_period('M')
df_clean
```

```python
# Calculate total revenue per month
top_products_month = df_clean.groupby('Month')['Revenue'].sum()
top_products_month
```

```python
# Plot revenue per month
sn.set_context("notebook", rc={"xtick.labelsize": 15, "ytick.labelsize": 15})

top_products_month = top_products_month.sort_index()

labels_X = top_products_month.index
values_Y = top_products_month.values

top_dates_df = pd.DataFrame({'Month': labels_X, 'Revenue': values_Y})

plt.figure(figsize=(15, 6))
plt.subplot(111)

sn.pointplot(data=top_dates_df, x='Month', y='Revenue')

plt.xticks(rotation=90)
plt.ylabel("Revenue")
plt.title("Months Generating the Most Revenue")

plt.show()
```

Monthly Trend 

Revenue varies significantly across months, with noticeable peaks during certain periods.
This suggests seasonal demand patterns.
The company should anticipate high-demand months by increasing inventory and launching targeted marketing campaigns in advance.

```python
# Top 20 countries by revenue
revenue_by_country = df_clean.groupby('Country')['Revenue'].sum().sort_values(ascending=False)[:20]
revenue_by_country.head(20)
```

```python
# Plot Top 20 countries by revenue
plt.figure(figsize=(15,6))
plt.subplot(111)

# Get labels and values
labels_X = revenue_by_country.index
values_Y = revenue_by_country.values

plt.bar(labels_X, values_Y)
plt.xticks(rotation=90)
plt.yscale("log")

# Add labels and title
plt.ylabel("Total Revenue")
plt.title("Top 20 Countries Generating the Most Revenue")

plt.show()
```

Revenue by Country

Revenue is highly concentrated in a limited number of countries, with one dominant market generating the majority of sales.
This creates geographic dependency risk.
Expanding operations and marketing efforts in other high-potential countries could help diversify revenue streams and reduce business risk.

This analysis highlights that revenue is concentrated on a limited number of high-performing products and primarily generated from one dominant country.
Sales also exhibit seasonal patterns, with revenue peaks during specific months.
Strategic recommendations include prioritizing top-performing products, optimizing seasonal inventory planning, and expanding geographically to diversify revenue sources.
